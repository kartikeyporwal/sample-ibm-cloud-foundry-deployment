```python
import numpy as np

x = np.arange(-5.0, 5.0, 0.1)
y = 2*(x) + 3
y_noise = 2 * np.random.normal(size=x.size)
ydata = y + y_noise
```

```python
import pandas as pd

df = pd.DataFrame()
df['Independent Variable'] = x
df['Dependent Variable'] = ydata
df.head()
```

```python
df.to_csv('data.csv', sep=',')
```
